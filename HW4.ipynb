```python
import pandas as pd
import os
file_path_1 = os.path.join("Resources","purchase_data.json")
file_path_2 = os.path.join("Resources","purchase_data.json")
```

```python
prch_data_1_df = pd.read_json(path_or_buf=file_path_1)
prch_data_2_df = pd.read_json(path_or_buf=file_path_2)
merged_df = prch_data_1_df.append(prch_data_2_df)
merged_df.reset_index(inplace=True)
merged_df
```

```python
total_players = merged_df["SN"].nunique()
total_players
```

```python
#Purchasing Analysis (Total)
#Number of Unique Items
#Average Purchase Price
#Total Number of Purchases
#Total Revenue
no_of_unique_items = merged_df["Item ID"].nunique()
print(f"Total number of unique items {no_of_unique_items}")
avg_purchase_price = merged_df["Price"].mean()
print("Average purchase price " '${:,.2f}'.format(avg_purchase_price))
total_revenue = merged_df["Price"].sum()
print("Total Revenue " '${:,.2f}'.format(total_revenue))
```

```python
#Gender Demographics

#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed
total_count = pd.to_numeric(total_players)
male_count = pd.to_numeric(merged_df[merged_df["Gender"] == 'Male']["SN"].nunique())
male_percentage = male_count*100/total_count
female_count =  pd.to_numeric(merged_df[merged_df["Gender"] == 'Female']["SN"].nunique())
female_percentage = female_count*100/total_count
other_count = total_count - male_count - female_count
other_percentage = other_count*100/total_count
print(f"Number of male counts is {male_count} and percentage wise it is {male_percentage}")
print(f"Number of female counts is {female_count} and percentage wise it is {female_percentage}")
print(f"Number of other counts is {other_count} and percentage wise it is {other_percentage}")
```

```python
#Purchasing Analysis (Gender)

#The below each broken by gender
#Purchase Count
#Average Purchase Price
#Total Purchase Value
#Normalized Totals
total_purchase =  merged_df["Price"].sum()
male_df = merged_df[merged_df["Gender"] == 'Male']
male_purchase_count = male_df["Gender"].count()
male_avg_purchase =  male_df["Price"].mean()
male_total_purchase = male_df["Price"].sum()
male_normal_purchase = male_total_purchase/male_count
print("-------------------MALE PURCHASE-----------")
print(f"Total purchase counts by male players {male_purchase_count}")
print("Average purchase price for male players "'${:,.2f}'.format(male_avg_purchase))
print("Total purchase price by male players "'${:,.2f}'.format(male_total_purchase))
print("Number of dollars spent by a single male player is "'${:,.2f}'.format(male_normal_purchase))

print("-------------------FEMALE PURCHASE-----------")
female_df = merged_df[merged_df["Gender"] == 'Female']
female_purchase_count = female_df["Gender"].count()
female_avg_purchase =  female_df["Price"].mean()
female_total_purchase = female_df["Price"].sum()
female_normal_purchase = female_total_purchase/female_count
print(f"Total purchase counts byfe male players {female_purchase_count}")
print("Average purchase price for female players "'${:,.2f}'.format(female_avg_purchase))
print("Total purchase price by female players "'${:,.2f}'.format(female_total_purchase))
print("Number of dollars spent by a single female player is "'${:,.2f}'.format(female_normal_purchase))

print("-------------------OTHER PURCHASE-----------")
other_df = merged_df[(merged_df["Gender"] != 'Female')  & (merged_df["Gender"] != 'Male')]
other_purchase_count = other_df["Gender"].count()
other_avg_purchase =  other_df["Price"].mean()
other_total_purchase = other_df["Price"].sum()
other_normal_purchase = other_total_purchase/other_count
print(f"Total purchase counts by other players {other_purchase_count}")
print("Average purchase price for other players "'${:,.2f}'.format(other_avg_purchase))
print("Total purchase price by other players "'${:,.2f}'.format(other_total_purchase))
print("Number of dollars spent by a single other player is "'${:,.2f}'.format(other_normal_purchase))
```

```python
#Age Demographics

#The below each broken into bins of 4 years (i.e. <10, 10-14, 15-19, etc.)
##Purchase Count
#Average Purchase Price
#Total Purchase Value
#Normalized Totals
bins=[0,10,14,19,24,29,34,39,44,49,54]
# Create labels for these bins
group_labels = ["<10","10-14","15-19","20-24","25-29","30-34",
                "35-39","40-44","45-49",">=75"]

# Slice the data and place it into bins
pd.cut(pd.to_numeric(merged_df["Age"]),bins,labels=group_labels).head()

# Place the data series into a new column inside of the DataFrame
merged_df["Age Group"] = pd.cut(merged_df["Age"],bins,labels=group_labels)
grouped_by_age_groups = merged_df.groupby(['Age Group'])
group_by_age_summary = pd.DataFrame({"Purchase Count":grouped_by_age_groups["Age"].count(),
                                   "Average Purchase Price":grouped_by_age_groups["Price"].mean(),
                                "Total Purchase Value":grouped_by_age_groups["Price"].sum(),"Normalized Totals":grouped_by_age_groups["Price"].sum()/total_purchase})
group_by_age_summary.reset_index(inplace=True)
group_by_age_summary.dropna(how="any")
```

```python
group_by_username = merged_df.groupby(['SN'])

group_by_username["Price"].sum()
total_spender_summary = pd.DataFrame(group_by_username["Price"].sum())
renamed_spender_summary = total_spender_summary.rename(columns={"Price":"Total Purchase Value"})
# Sort by strongest Pokemon
highest_spender = renamed_spender_summary.sort_values(["Total Purchase Value"], ascending=False)

highest_spender["Purchase Count"] = pd.DataFrame(group_by_username["SN"].count())

highest_spender["Average Purchase Price"] = pd.DataFrame(group_by_username["Price"].mean())
highest_spender
```
